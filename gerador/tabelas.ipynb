```python
import pandas as pd
import gg
u = 1
pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 10)
```

```python
ct = pd.read_csv("data/catalogo.csv",encoding="UTF-8",encoding_errors="replace")
ct = ct.loc[ct["ISBN"].notnull()]
ct = ct.loc[ct["Material type"] == "Book"]
```

```python
b = ct[["ISBN", "Title","Date of Publication","Name of Publisher","Authors"]].sample(n = 100).reset_index(drop=True)
fn = pd.read_csv("input/pnomes.csv",encoding="UTF-8",encoding_errors="replace")["pnome"]
ln = pd.read_csv("input/snomes.csv",encoding="UTF-8",encoding_errors="replace")["snome"]
un = pd.read_csv("input/usuarios.csv",encoding="UTF-8",encoding_errors="replace")["user"]
cc = pd.read_csv("input/cc.csv",encoding="UTF-8",encoding_errors="replace")["alpha-2"]
cu = pd.read_csv("input/cursos.csv",encoding="UTF-8",encoding_errors="replace")["curso"]
ca = pd.read_csv("input/categorias.csv",encoding="UTF-8",encoding_errors="replace")["categoria"]
```

```python
c = pd.DataFrame(cu.sample(n = 10)).rename(columns={"curso" : "NomeCurso"}).reset_index(drop=True)
c
```

<h1>cursos<h1>

```python
p = gg.professor(u,fn,ln)
u += len(p.index)
p
```

<h1>Professores<h1>

```python
a = gg.aluno(u,fn,ln)
u += len(a.index)
a
```

<h1>Alunos<h1>

```python
at = gg.telefones("Alunos_Matricula",a)
at
```

<h1>Telefones Alunos<h1>

```python
f = gg.funcionario(u,fn,ln)
u += len(f.index)
f
```

<h1>Funcionarios<h1>

```python
ft = gg.telefones("Funcionarios_Matricula",a)
ft
```

<h1>Telefones Funcionarios<h1>

```python
ut = gg.usuario(len(a.index),len(p.index),len(f.index),un)
ut
```

<h1>Usuarios<h1>

```python
livros = b[["ISBN", "Title","Date of Publication","Name of Publisher"]].rename(columns={"Title" : "Titulo", "Date of Publication" : "AnoDeLancamento", "Name of Publisher" : "Editora"})
livros = livros.assign(QuanCopias = gg.rc(pd.DataFrame(range(3,10), columns=["codes"])["codes"],livros["ISBN"]),
                       Categoria_CodCategoria = gg.rc(pd.DataFrame(range(1,5), columns=["codes"])["codes"],livros["ISBN"]))
livros
```

<h1>Livros<h1>

```python
cat = pd.DataFrame(ca.sample(n = 10)).rename(columns={"categoria" : "descricao"}).reset_index(drop=True)
cat
```

<h1>Categorias<h1>

```python
autores = b[["Authors"]].fillna("")
autores = autores.assign(Authors = autores["Authors"].str.split(";")).explode("Authors").drop_duplicates(keep="last").reset_index(drop=True)
autores = autores.assign(Authors = autores["Authors"].str.replace(" ","").str.replace(",",", "))
autores = autores.assign(Email = gg.email(15,autores["Authors"]),
                         Nacionalidade = gg.rc(cc,autores["Authors"])).\
                         rename(columns={"Authors" : "Nome"})
autores
```

<h1>Autores<h1>

```python
ral = b[["Authors","ISBN"]].fillna("").rename(columns={"Authors" : "Nome"})
ral = ral.assign(Nome = ral["Nome"].str.split(";")).explode("Nome").reset_index(drop=True)
ral = ral.assign(Nome = ral["Nome"].str.replace(" ","").str.replace(",",", "))
ral = pd.merge(autores, ral)[["Email","ISBN"]].rename(columns={"Email" : "Autores_Email","ISBN" : "Livros_ISBN"})
ral
```
