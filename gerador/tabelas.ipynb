```python
import pandas as pd
import gg
u = 1
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 10)
```

```python
ct = pd.read_csv("data/catalogo.csv",encoding='UTF-8',encoding_errors='replace')
ct = ct.loc[ct['ISBN'].notnull()]
ct = ct.loc[ct['Material type'] == "Book"]
```

```python
b = ct[["ISBN", "Title","Date of Publication","Name of Publisher","Authors"]].head(2000).reset_index(drop=True)
fn = pd.read_csv("data/pnomes.csv",encoding='UTF-8',encoding_errors='replace')["pnome"]
ln = pd.read_csv("data/snomes.csv",encoding='UTF-8',encoding_errors='replace')["snome"]
un = pd.read_csv("data/usuarios.csv",encoding='UTF-8',encoding_errors='replace')["user"]
```

```python
p = gg.professor(u,fn,ln)
u += len(p.index)
p
```
